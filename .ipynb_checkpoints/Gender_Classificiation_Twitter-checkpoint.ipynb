```python
# convert hex string to rgb ratios
import webcolors

def hex_to_rgb(hex_code):
    hex_code = '#' + hex_code 
    try:
        rgb = webcolors.hex_to_rgb(hex_code)
    except ValueError:
        hex_code = "#0084B4"
        rgb = webcolors.hex_to_rgb(hex_code)
    sum_rgb = sum(rgb)
    ratios = [comp / sum_rgb for comp in rgb]
    return ratios

# columns for the RGB ratios
df_dropped['red_ratio'], df_dropped['green_ratio'], df_dropped['blue_ratio'] = zip(*df_dropped['link_color'].apply(hex_to_rgb))
print(df_dropped[['link_color', 'red_ratio', 'green_ratio', 'blue_ratio']])
```

### Preprocessing texts for wordclouds: 
1) Removing punctuations, integers, and stopwords
2) Tokenize
3) Lemmatize

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    # remove integers
    text = ''.join([i for i in text if not i.isdigit()])
    # remove punctuations
    text = text.translate(str.maketrans('', '', string.punctuation))
    # tokenize text
    tokens = word_tokenize(text)
    # removing stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token.lower() not in stop_words]
    # lemmatize tokens
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    
    return ' '.join(tokens)

from tqdm import tqdm

# Running this line will avoid tqdm printing multiple lines in Jupyter Lab
tqdm._instances.clear()

pbar = tqdm(total=len(df_dropped)*3,desc="Preprocessing Text")

def preprocess_with_progress(text):
    cleaned_text = preprocess_text(text)
    pbar.update(1)
    
    return cleaned_text

df_dropped['cleaned_description'] = df_dropped['description'].apply(preprocess_with_progress)
df_dropped['cleaned_text'] = df_dropped['text'].apply(preprocess_with_progress)
df_dropped['cleaned_named'] = df_dropped['name'].apply(preprocess_with_progress)

pbar.close()
```

```python
import tensorflow as tf
from transformers import TFDistilBertModel, DistilBertTokenizer, DistilBertForSequenceClassification
```

```python
train_data = df_train[['gender', 'description']].copy()
val_data = df_val[['gender', 'description']].copy()
test_data = df_test[['gender', 'description']].copy()

train_texts = train_data['description'].tolist()
train_labels = train_data['gender'].tolist()

val_texts = val_data['description'].tolist()
val_labels = val_data['gender'].tolist()

test_texts = test_data['description'].tolist()
test_labels = test_data['gender'].tolist()

tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')

# tokenize the text
train_encodings = tokenizer(train_texts, truncation=True, padding=True, return_tensors='pt',
    max_length=128)
test_encodings = tokenizer(test_texts, truncation=True, padding=True, return_tensors='pt', 
    max_length=128)
val_encodings = tokenizer(val_texts, truncation=True, padding=True, return_tensors='pt', 
    max_length=128)
```

```python
# Adjusted from: https://towardsdatascience.com/multi-class-text-classification-with-deep-learning-using-bert-b59ca2f5c613
import torch
from torch.utils.data import TensorDataset

input_ids_train = train_encodings['input_ids']
attention_masks_train = train_encodings['attention_mask']
labels_train = torch.tensor(train_labels)

input_ids_val = val_encodings['input_ids']
attention_masks_val = val_encodings['attention_mask']
labels_val = torch.tensor(val_labels)

input_ids_test = test_encodings['input_ids']
attention_masks_test = test_encodings['attention_mask']
labels_test = torch.tensor(test_labels)

dataset_train = TensorDataset(input_ids_train, attention_masks_train, labels_train)
dataset_val = TensorDataset(input_ids_val, attention_masks_val, labels_val)
dataset_test = TensorDataset(input_ids_test, attention_masks_test, labels_test)
```

```python
model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased",
                                                      num_labels=3,
                                                      output_attentions=False,
                                                      output_hidden_states=False)
```

```python
# Adjusted from: https://towardsdatascience.com/multi-class-text-classification-with-deep-learning-using-bert-b59ca2f5c613
from torch.utils.data import DataLoader, RandomSampler, SequentialSampler

batch_size = 4

dataloader_train = DataLoader(dataset_train, 
                              sampler=RandomSampler(dataset_train), 
                              batch_size=batch_size)

dataloader_validation = DataLoader(dataset_val, 
                                   sampler=SequentialSampler(dataset_val), 
                                   batch_size=batch_size)

dataloader_test = DataLoader(dataset_test, 
                                   sampler=SequentialSampler(dataset_test), 
                                   batch_size=batch_size)
```

```python
# Adjusted from: https://towardsdatascience.com/multi-class-text-classification-with-deep-learning-using-bert-b59ca2f5c613
from transformers import AdamW, get_linear_schedule_with_warmup

optimizer = AdamW(model.parameters(),
                  lr=1e-5, 
                  eps=1e-8)

epochs = 5

scheduler = get_linear_schedule_with_warmup(optimizer, 
                                            num_warmup_steps=200,
                                            num_training_steps=len(dataloader_train)*epochs)
```

```python
# Adjusted from: https://towardsdatascience.com/multi-class-text-classification-with-deep-learning-using-bert-b59ca2f5c613
from sklearn.metrics import f1_score

def f1_score_func(preds, labels):
    preds_flat = np.argmax(preds, axis=1).flatten()
    labels_flat = labels.flatten()
    return f1_score(labels_flat, preds_flat, average='weighted')

def accuracy_per_class(preds, labels):
    label_dict_inverse = {v: k for k, v in label_dict.items()}
    
    preds_flat = np.argmax(preds, axis=1).flatten()
    labels_flat = labels.flatten()

    for label in np.unique(labels_flat):
        y_preds = preds_flat[labels_flat==label]
        y_true = labels_flat[labels_flat==label]
        print(f'Class: {label_dict_inverse[label]}')
        print(f'Accuracy: {len(y_preds[y_preds==label])}/{len(y_true)}\n')
```

```python
#  smaller subset of data for testing
num_samples = 2000

small_train_dataset = TensorDataset(
    input_ids_train[:num_samples],
    attention_masks_train[:num_samples],
    labels_train[:num_samples]
)


small_val_dataset = TensorDataset(
    input_ids_val[:num_samples],
    attention_masks_val[:num_samples],
    labels_val[:num_samples]
)

dataloader_train = DataLoader(small_train_dataset, 
                              sampler=RandomSampler(small_train_dataset), 
                              batch_size=batch_size)

dataloader_validation = DataLoader(small_val_dataset, 
                                   sampler=SequentialSampler(small_val_dataset), 
                                   batch_size=batch_size)
```

```python
def evaluate(dataloader_val):

    model.eval()
    
    loss_val_total = 0
    predictions, true_vals = [], []
    
    for batch in dataloader_val:
        
        batch = tuple(b.to(device) for b in batch)
        
        inputs = {'input_ids':      batch[0],
                  'attention_mask': batch[1],
                  'labels':         batch[2],
                 }

        with torch.no_grad():        
            outputs = model(**inputs)
            
        loss = outputs[0]
        logits = outputs[1]
        loss_val_total += loss.item()

        logits = logits.detach().cpu().numpy()
        label_ids = inputs['labels'].cpu().numpy()
        predictions.append(logits)
        true_vals.append(label_ids)
    
    loss_val_avg = loss_val_total/len(dataloader_val) 
    
    predictions = np.concatenate(predictions, axis=0)
    true_vals = np.concatenate(true_vals, axis=0)
            
    return loss_val_avg, predictions, true_vals
```

```python
# Adjusted from: https://towardsdatascience.com/multi-class-text-classification-with-deep-learning-using-bert-b59ca2f5c613
import random

seed_val = 42
random.seed(seed_val)
np.random.seed(seed_val)
torch.manual_seed(seed_val)
torch.cuda.manual_seed_all(seed_val)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

for epoch in range(1, epochs+1):
    model.train()
    loss_train_total = 0
    training_loss = []

    progress_bar = tqdm(dataloader_train, desc=f'Epoch {epoch}', leave=False)
    for batch in progress_bar:
        model.zero_grad()
        batch = tuple(b.to(device) for b in batch)
        inputs = {
            'input_ids': batch[0],
            'attention_mask': batch[1],
            'labels': batch[2]
        }
        outputs = model(**inputs)
        loss = outputs[0]
        loss_train_total += loss.item()
        training_loss.append(loss.item())
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()
        progress_bar.set_postfix({'Training loss': '{:.3f}'.format(loss.item() / len(batch))})

    torch.save(model.state_dict(), f'BERT_models/finetuned_BERT_epoch_{epoch}.model')
    loss_train_avg = loss_train_total / len(dataloader_train)
    tqdm.write(f'Epoch {epoch}')
    tqdm.write(f'Training loss: {loss_train_avg}')

    val_loss, predictions, true_vals = evaluate(dataloader_validation)
    val_f1 = f1_score_func(predictions, true_vals)
    tqdm.write(f'Validation loss: {val_loss}')
    tqdm.write(f'F1 Score (Weighted): {val_f1}')
```

```python
# new feature: uppercase count
df_train['uppercase_count'] = df_train['name'].str.count(r'[A-Z]')
df_val['uppercase_count'] = df_val['name'].str.count(r'[A-Z]')
df_test['uppercase_count'] = df_test['name'].str.count(r'[A-Z]')

# remove non alphanumeric and lowercase
df_train['name_processed'] = df_train['name'].replace('[^a-zA-Z0-9]', '', regex=True).str.lower()
df_val['name_processed'] = df_val['name'].replace('[^a-zA-Z0-9]', '', regex=True).str.lower()
df_test['name_processed'] = df_test['name'].replace('[^a-zA-Z0-9]', '', regex=True).str.lower()
```

```python
vectorizer = CountVectorizer(analyzer='char')
X_train_boc = vectorizer.fit_transform(df_train['name_processed'])
X_val_boc = vectorizer.transform(df_val['name_processed'])
X_test_boc = vectorizer.transform(df_test['name_processed'])
len(vectorizer.vocabulary_)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
from keras.models import Sequential
from keras.layers import Embedding, Conv1D, MaxPooling1D, LSTM, Dense, Flatten

vocab_size = len(vectorizer.vocabulary_)
max_sequence_length = max(len(name) for name in df_train['name_processed'])
classifiers = {
    'Naive Bayes': (MultinomialNB(), {'alpha': [0.1, 0.5, 1]}),
    'Gradient Boosting': (HistGradientBoostingClassifier(max_iter=75), {'max_depth' : [None,10]})
    
}


val_predictions = {}
for clf_name, (clf, param_grid) in classifiers.items():
    print(f"Training {clf_name}...")
    
    if isinstance(clf, HistGradientBoostingClassifier):
        clf_input = X_train_boc.toarray()
        val_input = X_val_boc.toarray()
    else:
        clf_input = X_train_boc
        val_input = X_val_boc
    
    grid_search = GridSearchCV(clf, param_grid, cv=5, scoring='accuracy')
    grid_search.fit(clf_input, df_train['gender'])
    
    best_clf = grid_search.best_estimator_
    print("Best Params:",best_clf)
    
    val_predictions[clf_name] = best_clf.predict(val_input)
    
    print(f"Evaluation for {clf_name}:")
    print(classification_report(df_val['gender'], val_predictions[clf_name]))
```

```python
from sklearn.metrics import classification_report
from keras.models import Sequential
from keras.layers import Embedding, Conv1D, MaxPooling1D, LSTM, Dense, Flatten
from keras.utils import pad_sequences
from keras.callbacks import EarlyStopping, ReduceLROnPlateau

epochs = 200
num_classes = 3
y_train = df_train['gender']
y_val = df_val['gender']
y_test = df_test['gender']

# Pad sequences to have the same length
max_sequence_length = max(len(name) for name in df_train['name_processed'])
X_train_padded = pad_sequences(X_train_boc.toarray(), maxlen=max_sequence_length)
X_val_padded = pad_sequences(X_val_boc.toarray(), maxlen=max_sequence_length)
X_test_padded = pad_sequences(X_test_boc.toarray(), maxlen=max_sequence_length)

# EarlyStopping callback
early_stopping = EarlyStopping(patience=10, restore_best_weights=True)

# ReduceLROnPlateau callback
reduce_lr = ReduceLROnPlateau(factor=0.5, patience=5)

# CNN model
cnn_model = Sequential()
cnn_model.add(Embedding(input_dim=X_train_padded.shape[1], output_dim=128, input_length=max_sequence_length))
cnn_model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
cnn_model.add(MaxPooling1D(pool_size=2))
cnn_model.add(Conv1D(filters=128, kernel_size=3, activation='relu'))
cnn_model.add(MaxPooling1D(pool_size=2))
cnn_model.add(Flatten())
cnn_model.add(Dense(256, activation='relu'))
cnn_model.add(Dense(num_classes, activation='softmax'))
cnn_model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

cnn_model.fit(X_train_padded, y_train, validation_data=(X_val_padded, y_val), epochs=epochs, batch_size=32,
             callbacks=[early_stopping, reduce_lr])

# Evaluate
cnn_probabilities = cnn_model.predict(X_test_padded)
cnn_predictions = np.argmax(cnn_probabilities, axis=1)
cnn_report = classification_report(y_test, cnn_predictions)
print("CNN Classification Report:")
print(cnn_report)

# RNN model
rnn_model = Sequential()
rnn_model.add(Embedding(input_dim=X_train_padded.shape[1], output_dim=128, input_length=max_sequence_length))
rnn_model.add(LSTM(units=128, return_sequences=True))
rnn_model.add(LSTM(units=64))
rnn_model.add(Dense(128, activation='relu'))
rnn_model.add(Dense(num_classes, activation='softmax'))
rnn_model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

rnn_model.fit(X_train_padded, y_train, validation_data=(X_val_padded, y_val), epochs=epochs, batch_size=32,
             callbacks=[early_stopping, reduce_lr])

# Evaluate
rnn_probabilities = rnn_model.predict(X_test_padded)
rnn_predictions = np.argmax(rnn_probabilities, axis=1)
rnn_report = classification_report(y_test, rnn_predictions)
print("RNN Classification Report:")
print(rnn_report)
```

```python
# Concatenate the "description" and "text" columns into a single corpus
desccorpus = df_train['description'] 
textcorpus = df_train['text']

# Define stopwords and lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Tokenize, lemmatize, and remove stopwords from the corpus
tokenized_desc = []
for text in desccorpus:
    tokens = gensim.utils.simple_preprocess(text)
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    filtered_tokens = [token for token in lemmatized_tokens if token not in stop_words]
    tokenized_desc.append(filtered_tokens)

tokenized_text = []
for text in textcorpus:
    tokens = gensim.utils.simple_preprocess(text)
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    filtered_tokens = [token for token in lemmatized_tokens if token not in stop_words]
    tokenized_text.append(filtered_tokens)

# Train the Word2Vec model
desc_w2v_model = Word2Vec(tokenized_desc, vector_size=100, window=5)
text_w2v_model = Word2Vec(tokenized_text, vector_size=100, window=5)

default_embedding = np.zeros(desc_w2v_model.vector_size)
desc_train = []
for tokens in tokenized_desc:
    desc_embeddings = [desc_w2v_model.wv[token] if token in desc_w2v_model.wv.key_to_index.keys() else default_embedding for token in tokens]
    if len(desc_embeddings) > 0:
        desc_train.append(np.mean(desc_embeddings, axis=0))  # Use the average of word embeddings for each text
    else:
        desc_train.append(default_embedding)
    
text_train = []
for tokens in tokenized_text:
    text_embeddings = [text_w2v_model.wv[token] if token in text_w2v_model.wv.key_to_index.keys() else default_embedding for token in tokens]
    if len(text_embeddings) > 0:
        text_train.append(np.mean(text_embeddings, axis=0))  # Use the average of word embeddings for each text
    else:
        text_train.append(default_embedding)
    
desc_train = np.array(desc_train)
text_train = np.array(text_train)
```

```python
def prepare_data(corpus, w2v_model, lemmatizer, stop_words):
    tokenized_corpus = []
    
    for text in corpus:
        tokens = gensim.utils.simple_preprocess(text)
        lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
        filtered_tokens = [token for token in lemmatized_tokens if token not in stop_words]
        tokenized_corpus.append(filtered_tokens)
    
    default_embedding = np.zeros(w2v_model.vector_size)  # for out-of-vocabulary words
    
    X = []
    for tokens in tokenized_corpus:
        embeddings = [w2v_model.wv[token] if token in w2v_model.wv else default_embedding for token in tokens]
        if len(embeddings) > 0:
            X.append(np.mean(embeddings, axis=0))  # average of word embeddings for each text
        else:
            X.append(default_embedding)
        
    return np.array(X, dtype=object)



val_desccorpus = df_val['description']
val_textcorpus = df_val['text']
X_val_desc = prepare_data(val_desccorpus, desc_w2v_model, lemmatizer, stop_words)
X_val_text = prepare_data(val_textcorpus, text_w2v_model, lemmatizer, stop_words)
y_val = df_val['gender'].values


test_desccorpus = df_test['description']
test_textcorpus = df_test['text']
X_test_desc = prepare_data(test_desccorpus, desc_w2v_model, lemmatizer, stop_words)
X_test_text = prepare_data(test_textcorpus, text_w2v_model, lemmatizer, stop_words)
y_test = df_test['gender'].values
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
import numpy as np

num_folds = 5

# Perform cross-validation with XGBoost
xgboost_scores = []
xgboost_models = []

skf = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)

fold = 1
for train_index, val_index in skf.split(final_df_train, final_train_y):
    print(f"Fold {fold}")
    X_train, X_val = final_df_train.iloc[train_index], final_df_train.iloc[val_index]
    y_train, y_val = final_train_y.iloc[train_index], final_train_y.iloc[val_index]

    xgboost_model = XGBClassifier(objective='multi:softmax', num_class=3)
    xgboost_model.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=0)

    xgboost_models.append(xgboost_model)
    score = accuracy_score(y_val, xgboost_model.predict(X_val))
    xgboost_scores.append(score)
    print(score)
    fold+=1

xgboost_mean_score = np.mean(xgboost_scores)
print(f"XGBoost Mean Accuracy: {xgboost_mean_score:.4f}")
```

```python
from catboost import CatBoostClassifier

num_folds = 5

catboost_scores = []
catboost_models = []

skf = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)

fold = 1
for train_index, val_index in skf.split(final_df_train, final_train_y):
    print(f"Fold {fold}")
    X_train, X_val = final_df_train.iloc[train_index], final_df_train.iloc[val_index]
    y_train, y_val = final_train_y.iloc[train_index], final_train_y.iloc[val_index]

    catboost_model = CatBoostClassifier(loss_function='MultiClass')
    catboost_model.fit(X_train, y_train, eval_set=(X_val, y_val), verbose=False, use_best_model=True)

    catboost_models.append(catboost_model)
    score = accuracy_score(y_val, catboost_model.predict(X_val))
    catboost_scores.append(score)
    print(score)
    fold += 1

catboost_mean_score = np.mean(catboost_scores)
print(f"CatBoost Mean Accuracy: {catboost_mean_score:.4f}")
```
